```python
a=int(input("enter the number to find its factorial"))
s=1
while(a!=0):
    s=s*a
    a=a-1
print(s) 
```

```python
c=int(input("enter the number to check whether armstrong or not "))
sum=0
n=c
while(c!=0):
    r=int(c%10)
    sum=sum+r**3
    c=int(c/10)
print(sum)
if(sum==n):
    print("armstrong no.")
else:
    print("not an armstrong no.")
```

```python
def check_armstrong(c):
    sum=0
    n=c
    while(c!=0):
        r=int(c%10)
        sum=sum+r**3
        c=int(c/10)
    if(sum==n):
        print("{} is armstrong no.".format(n))
    else:
        print("{} is not an armstrong no.".format(n))

print("enter the upper limit and lower limit of range to check whether any number is armstrong or not.")
a=int(input("a = "))
b=int(input("b = "))
for i in range(a,b+1):
    check_armstrong(i)
```
